##レシピ19.1　k-平均法によるクラスタリング

#問題

観測地をｋ個のグループにグループ分けしたい。

#解決策

k-平均法を用いる。

```
#ライブラリのロード
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

#データをロード
iris=datasets.load_iris()
features=iris.data[:,0:2]
```

```
iris.data
```

```
#特徴量を標準化
scaler=StandardScaler()
features_std=scaler.fit_transform(features)

#K-平均法クラスタリングを作成
cluster=KMeans(n_clusters=3,random_state=0,n_jobs=-1)

#k-平均法クラスタリングを訓練
model=cluster.fit(features_std)
```

```
iris.target_names  #ヒオウギアヤメ、ブルーフラック、バージニカ
```

```
#クラス予測値を表示
model.labels_
#0:'setosa', 1:'versicolor', 2:'virginica'
```

```
#クラスの真の値を表示
iris.target
```

```
from pandas import DataFrame

# 説明変数と目的変数を結合
df = DataFrame(features, columns=iris.feature_names)
df['target'] = iris.target_names[iris.target]
df
```

新たに観測値を予想させる

```
#新しい観測値
#sepal length (cm),sepal width (cm)	,petal length (cm),	petal width (cm)	
new_observation=[[3.6,1.5,5.5,0.2]]

model.predict(new_observation)
```

```
#クラスタの中心を表示
model.cluster_centers_
```

##レシピ19.2　k-平均法をクラスタリングの高速化

#問題
観測値をk個のグループにグループ分けしたいのだが、k-平均法では時間がかかりすぎる。

##解決策
ミニバッチk-平均法を用いる。


```
#ライブラリをロード
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MiniBatchKMeans

#データをロード
iris=datasets.load_iris()
features=iris.data

#特徴量を標準化
scaler=StandardScaler()
features_std=scaler.fit_transform(features)
```

```
#ミニバッチk-平均法クラスタリング器を作成
cluster=MiniBatchKMeans(n_clusters=3,random_state=0,batch_size=100)

#訓練
model=cluster.fit(features_std)
```

```
#クラス予測値を表示
model.labels_
```

```
#クラスの真の値を表示
iris.target
```

##レシピ19.3　平均値シフト法によるクラスタリング

#問題

クラスタの数や形を仮定せずに、観測値をグループ分けしたい。

#解決策

平均値シフト法クラスタリングを行う。


```
#ライブラリをロード
from sklearn.cluster import MeanShift
```

```
#平均値シフト法クラスタリング器を作成
cluster=MeanShift(n_jobs=-1)

#訓練
model=cluster.fit(features_std)
```

```
model.labels_
#0と1に分類された
```

##レシピ19.4　DBSCANを用いたクラスタリング

#問題

観測値を高密度のクラスタリングにグループ分けしたい。

#解決策

DBSCANクラスタリングを用いる。

```
#ライブラリをロード
from sklearn.cluster import DBSCAN
```

```
cluster=DBSCAN(n_jobs=-1)
model=cluster.fit(features_std)
```

```
print(model.labels_)
print(iris.target_names)
```
